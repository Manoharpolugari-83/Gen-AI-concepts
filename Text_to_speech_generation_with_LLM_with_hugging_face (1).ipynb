```python
from transformers import pipeline
```

```python
text = "My Name Is Polugari Manohar .My Favourite Course is Generative Ai , currently i am working in a asvix technologies role is chatbot developer intern "

pipe = pipeline("text-to-speech", model="suno/bark-small", device="cuda")

output = pipe(text)
```

```python
output
```

```python
from IPython.display import Audio

Audio(output["audio"], rate=output["sampling_rate"])
```
